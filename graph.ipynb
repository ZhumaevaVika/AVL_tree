```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

height = []
n = []
with open('height(n).txt', 'r') as file:
    for line in file:
        n.append(float(line.split()[0]))
        height.append(float(line.split()[1]) / 1000)

plt.scatter(n, height, s = 10, color = (0,0,1))

plt.title(r"График 1. height(n) (1000 запусков)")
plt.xlabel(r"n - количество элементов")
plt.ylabel(r"height - высота дерева")
plt.grid(True)
plt.minorticks_on()
plt.grid(which = 'minor', alpha = 0.2)
```
