```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

height = []
n = []
with open('height(n).txt', 'r') as file:
    for line in file:
        n.append(float(line.split()[0]))
        height.append(float(line.split()[1]) / 1000)

plt.scatter(n, height, s = 7, color = (0,0,1))

plt.title(r" 1. height(n) (1000 )")
plt.xlabel(r"n -  ")
plt.ylabel(r"height -  ")
plt.grid(True)
plt.minorticks_on()
plt.grid(which = 'minor', alpha = 0.2)

N = np.linspace(3, 200, 1000)
H = 1.72*np.log(N)
plt.plot(N, H, color = (1,0,0), label = r"$height = 1.72 \cdot log(n)$", linewidth = 2)
plt.legend()
```

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

n = []
time = []
with open('insert_time(n).txt', 'r') as file:
    for line in file:
        n.append(float(line.split()[0]))
        time.append(float(line.split()[1]))



plt.title(r" 2. time(n) (   )")
plt.xlabel(r"n -  ")
plt.ylabel(r"time - ")
plt.grid(True)
plt.minorticks_on()
plt.grid(which = 'minor', alpha = 0.2)

for i in range(len(n)):
    if time[i] > 140 or (time[i] < 99 and n[i] > 100):
        print(time[i], i)

time = np.concatenate((time[:2], time[3:16], time[17:23], time[24:26], time[27:42], time[43:48], time[49:]))
n = np.concatenate((n[:2], n[3:16], n[17:23], n[24:26], n[27:42], n[43:48], n[49:]))

plt.scatter(n, time, s = 7, color = (0,0,1))

N = np.linspace(3, 400, 1000)
H = 15*np.log2(N)
plt.plot(N, H, color = (1,0,0), label = r"$time = 1.72 \cdot log(n)$", linewidth = 2)
plt.legend()
```
